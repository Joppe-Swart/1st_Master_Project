```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('science')
from matplotlib import cm
from matplotlib.colors import ListedColormap
palette = cm.get_cmap("tab20", 20)
from astropy.io import fits


frequency = np.arange(0.1, 1000, 0.1)
nu_ff = 30
nu_dust = 200
S = 10
Synchrotron = S*(frequency/nu_ff)**(-0.8)
FF = 1.4*S*frequency**(-0.1)
Dust = 0.8*S*(frequency/nu_dust)**2
```

```python

plt.figure()
plt.axvline(x = 30, color = palette(1), alpha = 0.6)
plt.axvline(x = 200, color = palette(1))
plt.plot(frequency, FF, 'k--', alpha = 0.5, linewidth = 1)
plt.plot(frequency, Synchrotron, 'k-.', alpha = 0.5, linewidth = 1)
plt.plot(frequency, Dust, 'k:', alpha = 0.5, linewidth = 1)
plt.plot(frequency, FF + Synchrotron + Dust, 'k', linewidth = 1)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$\nu_{rest}$ [GHz]', fontsize = 12);
plt.ylabel(r'Luminosity [Arbitrary Units]', fontsize=12);
plt.xlim(1,900)
plt.ylim(1, 1100)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.axvspan(0.1, 30, facecolor = palette(1), alpha = 0.3)
plt.axvspan(30, 200, facecolor = palette(1), alpha = 0.6)
plt.axvspan(200, 950, facecolor = palette(1), alpha = 0.95)
plt.text(2, 500, 'Synchrotron', fontsize = 12)
plt.text(50, 500, 'FFE', fontsize = 12)
plt.text(250, 500, 'Dust', fontsize = 12)
plt.text(3, 250, r'$\propto \nu^{-0.8}$', fontsize = 12)
plt.text(50, 250, r'$\propto \nu^{-0.1}$', fontsize = 12)
plt.text(250, 250, r'$\propto \nu^{2}$', fontsize = 12)
plt.savefig('Characteristic Spectrum.pdf', dpi = 500)
plt.show()
```
